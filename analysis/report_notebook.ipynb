# Potentially inappropriate prescribing of DOACs to people with mechanical heart valves: a federated analysis of 57.9 million patientsâ€™ primary care records in situ using OpenSAFELY

In the early phase of the COVID-19 pandemic, NHS England recommended that, where appropriate, people prescribed warfarin should be switched to direct oral anticoagulants (DOACs) which require less frequent blood testing [[1](#reference_1)]. As warfarin requires frequent blood tests, this guidance was intended to support the reduction of pressures being experienced by the NHS and we have previously shown the increased switching of anticoagulants to DOACs following this guidance [[2](#reference_2)]. The use of DOACs in people with mechanical heart valves is not recommended and was explicitly advised against in the guidance issued by national bodies during the switching programme [[1](#reference_1),[3](#reference_3)]. Despite this, NHS England received anecdotal reports of those with mechanical heart valves being prescribed a DOAC as a result of switching anticoagulation during the pandemic. This report describes the monthly use of DOACs in people with mechanical heart valves throughout the COVID-19 pandemic in a federated analysis of 57.9 million people.
